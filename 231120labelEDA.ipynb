```python
import glob
import os

def glob_files(path, file_type = '*'):
    search_string = os.path.join(path, file_type)
    files = glob.glob(search_string)
    
    paths = []
    for f in files:
        if os.path.isdir(f):
            sub_paths = glob_files(f + '/')
            paths += sub_paths
        else:
            paths.append(f)
    
    paths.sort()

    return paths
```

만약 path가 폴더명이라면 뒤에 '/'를 붙여주고 다시 glob_files() 함수를 돌려라는 말.

```python
import numpy as np

WIDTH = 1080
HEIGHT = 1920

def load_labels(path):
    files = glob_files(path, '*.txt')

    Y_data = []
    for file in files:
        with open(file) as f:
            lines = f.readlines()

            boxes = []
            for line in lines:
                tokens = line.split()

                class_id = int(tokens[0])
                xc = float(tokens[1]) * WIDTH
                yc = float(tokens[2]) * HEIGHT
                width = float(tokens[3]) * WIDTH
                height = float(tokens[4]) * HEIGHT

                boxes.append(np.array([class_id, xc, yc, width, height]))

            Y_data.append(np.array(boxes))
        
    return Y_data
```

```python
basePath = 'highway/valid/label/'
Y_test = load_labels(os.path.join(basePath, 'Suwon_CH01_20200720_1830_MON_9m_RH_highway_TW5_sunny_FHD'))
```

```python
from glob import glob

glob(os.path.join(basePath, '*.xml'))
```

```python
valid_filename = []
for filepath in glob(os.path.join(basePath, '*.xml')):
    filename = os.path.basename(filepath)  ## 상위폴더 경로 제거
    filename = filename[:-4]  ## .xml 제거
    filename_parsed = filename.split('_')  ## 정보 쪼개기
    valid_filename.append(filename_parsed)  ## nested list

valid_filename
```

os.path.basename() 해도 확장자명은 따라온다!

```python
import pandas as pd

df = pd.DataFrame(valid_filename)
df.columns = ['region', 'channel', 'ymd', 'time', 'date', 'm', 'RushHour', 'road', 'twow', 'weather', 'quality']
df.head()
```
